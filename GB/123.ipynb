```python
import matplotlib.pyplot as plt
from sympy import *
import sympy
from math import sqrt
import numpy
#1.Определить корни
x=Symbol('x',domain=S.Reals)
fx=18*x**3+5*x**2 + 10*x-30
result=solveset(fx,x,Interval(-10,10))
print()
```

```python
#2.Найти интервалы, на которых функция возрастает
solve_univariate_inequality(fx>0,x)
```

```python
#3.Найти интервалы, на которых функция убывает
solve_univariate_inequality(fx<0,x)
```

```python
#4. Построить график
interval = numpy.arange(-100, 100,0.5)
fxlist=[fx.subs(x,value) for value in interval]
plt.plot(interval,fxlist)
plt.show()
```

```python
ydx=diff(18*x**3+5*x**2 + 10*x-30,x)
print(ydx)
resultydx=solveset(ydx,x,Interval(-10,10))
print(resultydx)
interval = numpy.arange(-100, 100,0.5)
ydxlist=[ydx.subs(x,value) for value in interval]
plt.plot(interval,ydxlist)
plt.show()
```

### Так как проме ydx >0 на промежутке (-оо;+oo), функция возрастает на промежутке (-оо;+oo)
